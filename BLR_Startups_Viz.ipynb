```python
import folium
import pandas as pd
import json
from folium import plugins

df = pd.read_csv('BLR Startup viz.csv')

with open('boundary.geojson') as f:
    blrArea = json.load(f)
```

```python
#group the dataframe by zip code and count the number of startups in each zip code
numStpSeries = df.groupby('PinCode').count().SN
#initialize an empty dataframe to store this new data
numStpsByZip = pd.DataFrame()
#populate the new dataframe with a 'zipcode' column and a 'numStartups' column
numStpsByZip['zipcode'] = [str(i) for i in numStpSeries.index]
numStpsByZip['numStartups'] = numStpSeries.values
```

```python
numStpsByZip['zipcode'].dtype
```

```python
numStpsByZip['zipcode'] = numStpsByZip['zipcode'].astype(int)
```

```python
numStpsByZip['zipcode'].dtype
```

```python
numStpsByZip.dtypes
```

```python
#initialize the BLR map
blrMap = folium.Map(location=[12.9720, 77.5939], tiles='cartodb positron', zoom_start=9)

blrMap.choropleth(geo_data="boundary.geojson",
             data=numStpsByZip,
             columns=['zipcode', 'numStartups'],
             key_on='feature.properties.pin_code',
             fill_color='BuPu', fill_opacity=0.7, line_opacity=0.5,
             legend_name='BLR Startups by Pincode')

# folium.Choropleth(
#     geo_data="boundary.geojson",
#     name="choropleth",
#     data=numStpsByZip,
#     columns=["zipcode", "numStartups"],
#     key_on="feature.properties.pin_code",
#     fill_color="YlGn",
#     fill_opacity=0.7,
#     line_opacity=0.2,
#     legend_name="BLR Startups by Pincode",
# ).add_to(blrMap)

blrMap.save('blrChoropleth.html')
```
